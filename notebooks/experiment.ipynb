```python
import torch.nn as nn
import torch.nn.functional as F
import torch

class LSTMGenerator(nn.Module):
    def __init__(self, vocab_size, embed_size=32, hidden_size=64, num_layers=2):
        super(LSTMGenerator, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embed_size)
        self.lstm = nn.LSTM(embed_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, vocab_size)

    def forward(self, x, hidden=None):
        x = self.embedding(x)
        out, hidden = self.lstm(x, hidden)
        out = self.fc(out)
        return out, hidden
```

```python
import torch
import torch.nn as nn
import os
import pickle

# Paths
MODEL_PATH = "../src/saved_models/best_lstm_model.pt"
ARTIFACTS_PATH = "../artifacts/preprocess_meta_lstm"

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")


def load_vocab(meta_path):
    """Load vocab dicts from preprocessing artifacts"""
    with open(meta_path, "rb") as f:
        meta = pickle.load(f)
    stoi, itos = meta["stoi"], meta["itos"]
    return stoi, itos


def generate_url(model, stoi, itos, max_len=100, temperature=1.0, seed="<s>"):
    """Generate a URL character by character"""
    model.eval()
    input_seq = torch.tensor([[stoi.get(c, stoi["<unk>"])] for c in seed], dtype=torch.long).T.to(DEVICE)

    hidden = None
    generated = [c for c in seed]

    for _ in range(max_len):
        with torch.no_grad():
            output, hidden = model(input_seq, hidden)
        
        # Get logits for the last character
        logits = output[:, -1, :] / temperature
        probs = torch.softmax(logits, dim=-1)

        next_id = torch.multinomial(probs, num_samples=1).item()
        next_char = itos[next_id]

        if next_char == "<eos>":
            break

        generated.append(next_char)

        # Feed back the predicted char
        input_seq = torch.tensor([[next_id]], dtype=torch.long).to(DEVICE)

    return "".join(generated)
```

```python
stoi, itos = load_vocab(ARTIFACTS_PATH)

# Load model
model = LSTMGenerator(vocab_size=len(stoi)).to(DEVICE)
model.load_state_dict(torch.load(MODEL_PATH, map_location=DEVICE))
model.eval()

# Generate a few URLs
for i in range(5):
    url = generate_url(model, stoi, itos, max_len=80, seed="https://")
    print(f"Generated URL {i+1}: {url}")
```
